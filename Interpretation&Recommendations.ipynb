Assign Labels and Group by Clusters

```python
# Add cluster labels to the original dataframe
customer_df['label'] = kmeans.labels_
print(customer_df)

# Group by cluster labels and calculate mean values
cluster_summary = customer_df.groupby('label').agg({'Age': 'mean', 
                                                    'Annual Income (k$)': 'mean', 
                                                    'Spending Score (1-100)': 'mean'})
print(cluster_summary)
```

Visualize Clusters in Original Feature Space

```python
# Scatter plot of clusters in the original feature space
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_df, x='Age', y='Spending Score (1-100)', hue='label', palette='Set1')
plt.title('Clusters in Original Feature Space')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
